## 1. Factorial of a number

```python
n = int(input("Enter number to find factorial: "))

fact = 1

for i in range(1, n+1):
    fact = fact * i

print(f"Factorial of {n} = {fact}")
```

## 2. Armstrong Number
( sum of cubes of digits = number itself )
153 → 1³ + 5³ + 3³ = 1 + 125 + 27 = 153

```python
num = int(input("Enter a number: "))
sum_of_cubes = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum_of_cubes += digit ** 3
    temp = temp // 10

if sum_of_cubes == num:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is NOT an Armstrong number")
```

## 3. Fibonacci series up to n terms

```python
n = int(input("How many terms? "))

a, b = 0, 1

if n <= 0:
    print("Please enter positive number")
elif n == 1:
    print(a)
else:
    print(a, end=' ')
    for i in range(2, n+1):
        print(b, end=' ')
        a, b = b, a + b
    print()   # new line
```

## 4. Prime number check

```python
n = int(input("Enter number: "))

if n < 2:
    print(f"{n} is NOT prime")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{n} is PRIME")
    else:
        print(f"{n} is NOT prime")
```

## 5. Palindrome number check

```python
num = int(input("Enter number: "))
original = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if original == rev:
    print("It is a Palindrome number")
else:
    print("It is NOT a Palindrome number")
```

## 6. Sum of digits

```python
n = int(input("Enter number: "))
sum_digits = 0

while n > 0:
    digit = n % 10
    sum_digits += digit
    n = n // 10

print("Sum of digits =", sum_digits)
```

## 7. Count numbers divisible by 5 or 7 between 1–100

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        count += 1
        print(i, end=' ')

print(f"\nTotal count = {count}")
```

## 8. Convert lowercase string to uppercase (without .upper())

```python
s = input("Enter any string: ")

result = ""

for char in s:
    if 'a' <= char <= 'z':
        result += chr(ord(char) - 32)
    else:
        result += char

print("Uppercase version:", result)
```

## 9. Print multiplication table of given number

```python
n = int(input("Enter number for table: "))

for i in range(1, 11):
    print(f"{n} × {i} = {n*i}")
```

## 10. Pattern (the diamond-like number pattern)

```python
n = 5   # you can change this

for i in range(1, n+1):
    # left numbers
    for j in range(1, n-i+2):
        print(j, end="")
    # stars
    for j in range(1, 2*i-1):
        print("*", end="")
    # right numbers (reverse)
    for j in range(n-i+1, 0, -1):
        print(j, end="")
    print()
```

## 11. Sum of series → 1 + 1/2 + 1/3 + ... + 1/n

```python
n = int(input("Enter n: "))

sum_series = 0.0

for i in range(1, n+1):
    sum_series += 1 / i

print(f"Sum = {sum_series:.6f}")
```
