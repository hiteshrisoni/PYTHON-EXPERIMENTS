### Difference Between Interactive Mode and Script Mode

Interactive Mode:
- Runs one line at a time.
- Shows output immediately.
- Used for testing small code.

Script Mode:
- Write full program in a file.
- Save the file.
- Run the program using Run option.
- Used for full programs.

## Question 8: Area of Triangle

```python
a = 3
b = 4
c = 5

s = (a + b + c) / 2
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

print("Area:", area)
```

## Question 9: Convert Seconds into Hours, Minutes and Seconds

```python
seconds = 3665

hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```
