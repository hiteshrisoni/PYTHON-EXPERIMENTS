## 1. Scan n values in range 0-3 and count frequency of each value (0,1,2,3)

```python
n = int(input("How many numbers will you enter? "))

counts = [0, 0, 0, 0]  # index 0→count of 0, index 1→count of 1, etc.

print("Enter", n, "numbers (only 0 to 3 allowed):")

for i in range(n):
    val = int(input(f"Number {i+1}: "))
    if 0 <= val <= 3:
        counts[val] += 1
    else:
        print("Warning: Only 0-3 allowed. Ignoring", val)

print("\nFrequency:")
for i in range(4):
    print(f"{i} appeared {counts[i]} time(s)")
```

## 2. Create a tuple of n numeric values and find their average

```python
n = int(input("How many numbers? "))

numbers = []

for i in range(n):
    val = float(input(f"Enter number {i+1}: "))
    numbers.append(val)

tup = tuple(numbers)           # convert list to tuple

if n > 0:
    avg = sum(tup) / len(tup)
    print("Tuple:", tup)
    print("Average:", round(avg, 2))
else:
    print("No numbers entered.")
```

## 3. Input N students' scores in a list. Find and print the runner-up score  
(Runner-up = second highest score)

```python
n = int(input("Enter number of students (N): "))

scores = []

print("Enter scores (space separated in one line):")
scores = list(map(int, input().split()))

if len(scores) != n:
    print("Warning: You entered", len(scores), "scores instead of", n)

# Remove duplicates and sort descending
unique_sorted = sorted(set(scores), reverse=True)

if len(unique_sorted) >= 2:
    print("Highest score:", unique_sorted[0])
    print("Runner-up score:", unique_sorted[1])
else:
    print("Not enough unique scores to find runner-up.")
```

## 4. Create dictionary of n persons → key: name, value: city  
Perform:  
a) Display all names  
b) Display all cities  
c) Display name and city of all  
d) Count students in each city

```python
n = int(input("How many persons? "))

students = {}

for i in range(n):
    name = input(f"Enter name {i+1}: ").strip()
    city = input(f"Enter city for {name}: ").strip()
    students[name] = city

print("\n--- Results ---")
print("a) All names:", list(students.keys()))

print("\nb) All cities:", list(students.values()))

print("\nc) Name → City")
for name, city in students.items():
    print(f"{name} → {city}")

# Count per city
from collections import Counter
city_count = Counter(students.values())

print("\nd) Students per city:")
for city, cnt in city_count.items():
    print(f"{city}: {cnt}")
```

## 5. Store movie details in a dictionary  
Each movie: name (key), value = {year, director, cost, collection}

```python
movies = {}

n = int(input("How many movies? "))

for i in range(n):
    name = input(f"\nMovie {i+1} name: ").strip()
    year = int(input("Year: "))
    director = input("Director: ").strip()
    cost = float(input("Production cost (in crores): "))
    earn = float(input("Collection/earning (in crores): "))

    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "collection": earn
    }

print("\n--- Movie Operations ---")

# a) Print all movie details
print("a) All movies:")
for name, info in movies.items():
    print(f"{name} ({info['year']}) - Dir: {info['director']}, "
          f"Cost: ₹{info['cost']} cr, Earned: ₹{info['collection']} cr")

# b) Movies before 2015
print("\nb) Movies released before 2015:")
for name, info in movies.items():
    if info["year"] < 2015:
        print(name)

# c) Movies that made profit
print("\nc) Movies that made profit:")
for name, info in movies.items():
    if info["collection"] > info["cost"]:
        print(f"{name} (Profit: ₹{info['collection'] - info['cost']} cr)")

# d) Movies by a particular director
director_name = input("\nd) Enter director name to search: ").strip()
print(f"Movies directed by {director_name}:")
found = False
for name, info in movies.items():
    if info["director"].lower() == director_name.lower():
        print(name)
        found = True
if not found:
    print("No movies found by this director.")
```

## 6. Simple Contact Book using Dictionary  
(Add, View, Update, Delete)

```python
contacts = {}

while True:
    print("\n--- Contact Book ---")
    print("1. Add contact")
    print("2. View all contacts")
    print("3. Search / Update contact")
    print("4. Delete contact")
    print("5. Exit")

    choice = input("Enter choice (1-5): ").strip()

    if choice == '1':
        name = input("Name: ").strip()
        phone = input("Phone: ").strip()
        contacts[name] = phone
        print("Contact added!")

    elif choice == '2':
        if contacts:
            print("All contacts:")
            for name, phone in contacts.items():
                print(f"{name}: {phone}")
        else:
            print("No contacts yet.")

    elif choice == '3':
        name = input("Enter name to search/update: ").strip()
        if name in contacts:
            print(f"Current: {name} → {contacts[name]}")
            new_phone = input("New phone (or press Enter to keep old): ").strip()
            if new_phone:
                contacts[name] = new_phone
            print("Updated!")
        else:
            print("Contact not found.")

    elif choice == '4':
        name = input("Enter name to delete: ").strip()
        if name in contacts:
            del contacts[name]
            print("Deleted.")
        else:
            print("Not found.")

    elif choice == '5':
        print("Goodbye!")
        break

    else:
        print("Invalid choice.")
```

## 7. To-Do List Manager using List  
(Add, View, Remove tasks)

```python
todo = []

while True:
    print("\n--- To-Do List ---")
    print("1. Add task")
    print("2. View all tasks")
    print("3. Remove task")
    print("4. Exit")

    ch = input("Choose (1-4): ").strip()

    if ch == '1':
        task = input("Enter task: ").strip()
        todo.append(task)
        print("Task added.")

    elif ch == '2':
        if todo:
            print("Your tasks:")
            for i, t in enumerate(todo, 1):
                print(f"{i}. {t}")
        else:
            print("No tasks yet.")

    elif ch == '3':
        if todo:
            print("Tasks:")
            for i, t in enumerate(todo, 1):
                print(f"{i}. {t}")
            idx = int(input("Enter task number to remove: ")) - 1
            if 0 <= idx < len(todo):
                removed = todo.pop(idx)
                print(f"Removed: {removed}")
            else:
                print("Invalid number.")
        else:
            print("Nothing to remove.")

    elif ch == '4':
        print("Exiting To-Do List. Have a great day!")
        break

    else:
        print("Wrong choice.")
```
