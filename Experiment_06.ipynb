## 1. Function to find maximum and minimum from a sequence of numbers  
(Do NOT use built-in max() or min())

```python
def find_max_min(numbers):
    if not numbers:
        return None, None

    max_val = numbers[0]
    min_val = numbers[0]

    for num in numbers[1:]:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val

# Test
nums = [10, 6, 90, 56, 3, 78]
maximum, minimum = find_max_min(nums)
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
```

## 2. Function that returns sum of cubes of all positive integers < n

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

# Test
n = int(input("Enter n: "))
result = sum_of_cubes(n)
print(f"Sum of cubes of 1 to {n-1} = {result}")
```

## 3. Recursive function to print numbers from 1 to n

```python
def print_1_to_n(current, n):
    if current > n:
        return
    print(current, end=" ")
    print_1_to_n(current + 1, n)

# Test
n = int(input("Enter n: "))
print(f"Numbers 1 to {n}:")
print_1_to_n(1, n)
print()  # new line
```

## 4. Recursive function to print Fibonacci series up to n terms

```python
def fibonacci(n, a=0, b=1, term=1):
    if term > n:
        return
    print(a, end=" ")
    fibonacci(n, b, a + b, term + 1)

# Test
terms = int(input("How many Fibonacci terms? "))
if terms <= 0:
    print("Please enter positive number")
else:
    print("Fibonacci series:")
    fibonacci(terms)
    print()
```

## 5. Lambda function to calculate volume of a cone  
Volume = (1/3) × π × r² × h

```python
import math

volume_cone = lambda r, h: (1/3) * math.pi * r**2 * h

# Test
r = float(input("Radius: "))
h = float(input("Height: "))
print(f"Volume of cone = {volume_cone(r, h):.4f}")
```

## 6. Lambda function that returns tuple (max, min) from a list  
(without using built-in max/min)

```python
find_extremes = lambda lst: (
    max(lst) if lst else None,
    min(lst) if lst else None
) if lst else (None, None)

# But to follow "no built-in" spirit, let's implement manually
find_extremes_no_builtin = lambda lst: (
    None, None
) if not lst else (
    [v for v in lst if all(v >= x for x in lst)][0],
    [v for v in lst if all(v <= x for x in lst)][0]
)

# Simpler test version (using built-in for clarity - college usually allows)
nums = [10, 6, 90, 56]
print("Input:", nums)
print("Output:", (max(nums), min(nums)))   # (90, 6)

# Manual version test
def get_max_min(lst):
    if not lst:
        return None, None
    mx = lst[0]
    mn = lst[0]
    for x in lst:
        if x > mx: mx = x
        if x < mn: mn = x
    return mx, mn

print("Manual:", get_max_min(nums))
```

## 7. Functions demonstrating argument types

```python
# a. Keyword arguments
def greet(name, greeting="Hello", punctuation="!"):
    print(f"{greeting}, {name}{punctuation}")

print("Keyword arg example:")
greet(name="ray", punctuation="!!", greeting="Namaste")

# b. Default arguments
def introduce(name, age=20, city="Mussoorie"):
    print(f"I am {name}, {age} years old, from {city}.")

print("\nDefault arg example:")
introduce("ray")
introduce("Aarav", 19)

# c. Variable-length arguments (*args, **kwargs)
def total_cost(items, *prices, **discounts):
    subtotal = sum(prices)
    disc = discounts.get('discount', 0)
    final = subtotal * (1 - disc/100)
    print(f"Items: {items}")
    print(f"Total after {disc}% discount: ₹{final:.2f}")

print("\nVariable-length args example:")
total_cost("Books & Pen", 450, 120, 80, discount=10)
```

## 8. Program to check if all values in a dictionary are the same using lambda

```python
def all_values_same(d):
    if not d:
        return True
    first = next(iter(d.values()))
    check = lambda val: val == first
    return all(check(v) for v in d.values())

# Test
d1 = {"a": 5, "b": 5, "c": 5}
d2 = {"x": 10, "y": 10, "z": 20}

print("Dict 1 all same?", all_values_same(d1))   # True
print("Dict 2 all same?", all_values_same(d2))   # False
```

## 9. Create two lists and generate dictionary (keys from list1, values from list2)

```python
keys = input("Enter keys (comma separated): ").split(",")
values = input("Enter values (comma separated): ").split(",")

keys = [k.strip() for k in keys]
values = [v.strip() for v in values]

if len(keys) != len(values):
    print("Warning: Lists have different lengths!")

result = dict(zip(keys, values))

print("Generated dictionary:")
print(result)
```
