```python
## 1. Count and display number of capital letters in a given string
```

```python
s = input("Enter a string: ")

count_cap = 0

for char in s:
    if 'A' <= char <= 'Z':
        count_cap += 1

print("Number of capital letters:", count_cap)
```

## 2. Count total number of vowels in a given string

```python
s = input("Enter a string: ").lower()   # make case insensitive

vowels = 'aeiou'
count_v = 0

for char in s:
    if char in vowels:
        count_v += 1

print("Total vowels:", count_v)
```

## 3. Input a sentence and print words in separate lines

## 4. Count how many times a substring occurs in main string  
(Left to right traversal only — normal .count() behavior)

```python
main_str = input("Enter main string: ")
sub = input("Enter substring to find: ")

count = main_str.count(sub)

print(f"'{sub}' occurs {count} time(s) in the string")
```

## 5. Count occurrences of each alphabet (case insensitive)  
Display only those that appear ≥1 time

```python
s = input("Enter string with mixed case: ").lower()

freq = {}

for char in s:
    if char.isalpha():           # only letters
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1

print("Letter occurrences (case insensitive):")
for letter in sorted(freq.keys()):
    print(f"{letter.upper()} → {freq[letter]}")
```

## 6. Count number of unique words in a given sentence using sets

```python
sentence = input("Enter a sentence: ")

words = sentence.split()          # default split on whitespace

unique_words = set(words)         # set removes duplicates

print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

## 7. Create two sets s1 and s2 of fruits (user input)  
Find:  
a) Fruits only in s1 but not in s2  
b) Fruits only in both s1 and s2 (intersection)  
c) Count of all fruits in s1 and s2 (union)

```python
print("Enter fruits for set s1 (comma separated):")
s1 = set(input().split(","))

print("Enter fruits for set s2 (comma separated):")
s2 = set(input().split(","))

# Clean up extra spaces
s1 = {fruit.strip() for fruit in s1}
s2 = {fruit.strip() for fruit in s2}

print("\nResults:")
print("a) Fruits only in s1 (not in s2):", s1 - s2)
print("b) Fruits in both s1 and s2:", s1 & s2)
print("c) All unique fruits (s1 ∪ s2):", s1 | s2)
print("   Total count:", len(s1 | s2))
```

## 8. Take two given sets of fruits and apply set operations
S1 = {'Red', 'yellow', 'orange', 'blue'}
S2 = {'violet', 'blue', 'purple'}

```python
S1 = {'Red', 'yellow', 'orange', 'blue'}
S2 = {'violet', 'blue', 'purple'}

print("S1:", S1)
print("S2:", S2)
print()

print("Union (all fruits)          :", S1 | S2)
print("Intersection (common)       :", S1 & S2)
print("Difference (S1 - S2)        :", S1 - S2)
print("Difference (S2 - S1)        :", S2 - S1)
print("Symmetric difference        :", S1 ^ S2)
```
